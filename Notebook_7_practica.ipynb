10) Selecciona el maximo de datos faltantes. Calcula que porcentaje de los datos representan los datos faltantes.

```python
# Obtengo el maximo valor de datos faltantes y lo guardo en la variable masnan
datos_faltantes = expvida.isnull().sum()
masnan = datos_faltantes.max()

print("Máximo valor de datos faltantes ->", masnan)
```

```python
# Obtengo la cantidad de filas total y divido el maximo valor de datos faltantes por este numero
# y lo multiplico por 100
porcentaje_faltantes = (masnan / len(expvida)) * 100

print("Porcentaje de datos faltantes -> ", porcentaje_faltantes)
```

```python
# Imprimo el maximo porcentaje de datos faltantes. Uso f-strings.
print(f"Porcentaje de datos faltantes -> {porcentaje_faltantes:.2f}%")
```

14) ¿Cual es la expectativa de vida mas alta y mas baja? Usa la funcion de pandas y la funcion de NumPy y compara el resultado. ¿Que paises tienen esas expectativas de vida?

```python
# Maxima expectativa de vida usando numpy
maxima_expectativa_numpy = np.max(expvida['life_expectancy'])
```

```python
# Maxima expectativa de vida usando pandas
maxima_expectativa_pandas = expvida['life_expectancy'].max()
```

```python
# Minima expectativa de vida usando numpy
minima_expectativa_numpy = np.min(expvida['life_expectancy'])
```

```python
# Minima expectativa de vida usando pandas
minima_expectativa_pandas = expvida['life_expectancy'].min()
```

```python
# Muestro los resultados maximos
print("Máxima expectativa de vida numpy -> ", maxima_expectativa_numpy)
print("Máxima expectativa de vida pandas -> ", maxima_expectativa_pandas)
```

```python
# Muestro los resultados minimos
print("Mínima expectativa de vida numpy -> ", minima_expectativa_numpy)
print("Mínima expectativa de vida pandas -> ", minima_expectativa_pandas)
```

```python
# Filtro las columnas para mostrar solo las filas que tengan valor
# igual al maximo
promedio_por_pais = expvida.groupby('Country')['life_expectancy'].mean()
valores_maximo = promedio_por_pais == promedio_por_pais.max()
paises_maximo = promedio_por_pais[valores_maximo]

print(paises_maximo)
```

```python
# Filtro las columnas para mostrar solo las filas que tengan valor
# igual al minimo
promedio_por_pais = expvida.groupby('Country')['life_expectancy'].mean()
valores_minimo = promedio_por_pais == promedio_por_pais.min()
paises_minimo = promedio_por_pais[valores_minimo]

print(paises_minimo)
```
